```python
own = pd.get_dummies(df["ownsership"],drop_first=True)
```

```python

df.drop("transmission",axis=1,inplace=True)
```

```python
#create car age
df["year_they_man"] = df["manufacturing_year"].astype(int)
```

```python
df.drop("registration_year",axis=1,inplace=True)
```

```python
type_of_fuel = pd.get_dummies(df["fuel_type"],drop_first=True)
```

```python
df = pd.concat([df,type_of_fuel],axis=1)
```

```python
df.drop("fuel_type",axis=1,inplace=True)
```

```python
X = df[['seats', 'kms_driven', 'mileage(kmpl)', 'engine(cc)', 'max_power(bhp)',
       'torque(Nm)', 'First Owner', 'Second Owner',
       'Third Owner', 'Manual', 'age', 'Diesel', 'Petrol']]
```

```python
y = df['price(in lakhs)']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```

```python
sns.distplot((y_test-predictions))
```

```python
from sklearn import metrics
```

```python
metrics.mean_absolute_error(y_test,predictions)
```

```python
metrics.mean_squared_error(y_test,predictions)
```

```python
np.sqrt(metrics.mean_squared_error(y_test,predictions))
```
